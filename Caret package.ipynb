## Summary

The <i>caret</i> package which is the short for <i> Classification And REgression Traning </i> involves several functions to streamline the process for regresion and classification problems. Max Kuhn and several contrubiters has created this useful package. 


As a beginner, thanks to caret packagae, one can train multiple models with one easy format and try different combinations of parameters to find the best model. Also one can evalute perfomance of those variable to understand their effect on the model and caret helps one to pick the best model. 

Basically the tools that are involved in this package enables: 
- data splitting
- pre-processing
- feature selection
- model tuning using resampling
- variable importance estimation
- visualisation

Indeed there are several tools that enables other functionalities.

## Caret vs other CRAN Packages

## Implementation of caret package on third lab 

### Simple Splitting Based on the Outcome

By using createDataPartition one can easily split the data into train and test data set. When the y argument is a factor to this function, the random sampling will occur within each class

```r
library(caret)
library(data.table)
options(repr.matrix.max.rows=20, repr.matrix.max.cols=15) # for limiting the number of top and bottom rows of tables printed 
datapath <- "~/data_ad454"
```

```r
realty_data <- readRDS(sprintf("%s/rds/02_01_realty_data.rds", datapath))
```

```r
realty_data2 <- realty_data[, c("price", "brut_metrekare","kat_sayisi", "kira_getirisi")]
realty_data2 <- na.omit(realty_data2)
realty_data2
```

```r
set.seed(1999)
train_indices <- createDataPartition(realty_data2$kira_getirisi, p = .7, 
                                  list = FALSE, 
                                  times = 1)
head(train_indices)
```

```r
train_data <- realty_data2[ train_indices,]
test_data  <- realty_data2[-train_indices,]
```

### Model traning 

Another advantage of caret is that we can use same function which is train() to run all of the models. Train has a 23 attributes that accept several caret spesific arguments. 

```r
reg_model <- train(price ~ ., data = train_data, 
                 method = "lm")
reg_model
```

```r
predicted_train <- predict(reg_model,train_data)
predicted_test <- predict(reg_model, test_data)
actual_train <- train_data$price
actual_test <- test_data$price
```

## Performance Measurements

```r
model_dt <- data.table(partition = c("train", "test"),
                       R2 = c(R2(predicted_train, actual_train),
                                R2(predicted_test, actual_test)),
                        RMSE = c(RMSE(predicted_train, actual_train),
                                 RMSE(predicted_test, actual_test)),
                        MAE = c(MAE(predicted_train, actual_train),
                                MAE(predicted_test, actual_test))
                        )
model_dt
```
